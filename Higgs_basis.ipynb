```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as ppt
```

```python
cba, MH, m, MHp, Y2, MA, beta = sp.symbols('cba, MH, m, MHp, Y2, MA, beta', real=True) # cba is cos(beta-alpha), tan(beta) = v2/v1, Y2 is a cuadratic term and the rest are masses.
```

```python
Z1 = (1-cba**2)*(MH**2*cba**2/(1-cba**2)+m**2)
Z3 = 2*MHp**2-2*Y2
Z4 = 1/2*(2*cba**2-1)*(m**2-MH**2)+m**2/2+MA**2+MH**2/2-2*MHp**2
Z5 = 1/2*(2*cba**2-1)*(m**2-MH**2)+m**2/2-MA**2+MH**2/2
Z6 = sp.sqrt(1-cba**2)*cba*(m**2-MH**2)

Z345 = Z3 + Z4 + sp.re(Z5)

Z2 = 1/(32*sp.cos(beta)**3*sp.sin(beta)**3)*(-8*Z6*(sp.cos(2*beta)+sp.cos(6*beta))+sp.sin(2*beta)*(Z1+2*Z345)+sp.sin(6*beta)*(2*Z345-3*Z1))
Z7 = 1/(8*sp.cos(beta)**2*sp.sin(beta)**2)*(-3*sp.cos(4*beta)*Z6+sp.sin(4*beta)*(-Z1+Z3+Z4+Z5)-Z6)
```

```python
lambda_1 = Z1*sp.cos(beta)**4+Z2*sp.sin(beta)**4+1/2*Z345*sp.sin(2*beta)**2+2*sp.sin(2*beta)*(sp.cos(beta)**2*sp.re(Z6)+sp.sin(beta)**2*sp.re(Z7))
lambda_2 = Z1*sp.sin(beta)**4+Z2*sp.cos(beta)**4+1/2*Z345*sp.sin(2*beta)**2-2*sp.sin(2*beta)*(sp.sin(beta)**2*sp.re(Z6)+sp.cos(beta)**2*sp.re(Z7))
lambda_3 = Z3+1/4*sp.sin(2*beta)**2*(Z1+Z2-2*Z345)-sp.sin(2*beta)*sp.cos(2*beta)*sp.re(Z6-Z7)
lambda_4 = Z4+1/4*sp.sin(2*beta)**2*(Z1+Z2-2*Z345)-sp.sin(2*beta)*sp.cos(2*beta)*sp.re(Z6-Z7)
lambda_5 = 1/4*sp.sin(2*beta)**2*(Z1+Z2-2*Z345)+sp.re(Z5)-sp.sin(2*beta)*sp.cos(2*beta)*sp.re(Z6-Z7)
```

# Analysis

```python
zeta_1_example = Z1.subs({m:0, Y2:0, MHp:MH, MA:MH})  # This is an example
zeta_3_example = Z3.subs({m:0, Y2:0, MHp:MH, MA:MH})  # Example
zeta_4_example = Z4.subs({m:0, Y2:0, MHp:MH, MA:MH})  # Example
zeta_5_example = Z5.subs({m:0, Y2:0, MHp:MH, MA:MH})  # Example
zeta_6_example = Z6.subs({m:0, Y2:0, MHp:MH, MA:MH})  # Example

zeta_1_example = zeta_1_example.subs({MH:0.6/0.25, beta:np.pi/4}) # Example
zeta_1_num = sp.lambdify([cba], zeta_1_example, modules='numpy')

zeta_3_example = zeta_3_example.subs({MH:0.6/0.25, beta:np.pi/4}) # Example
zeta_3_num = np.linspace(zeta_3_example,zeta_3_example,400)

zeta_4_example = zeta_4_example.subs({MH:0.6/0.25, beta:np.pi/4}) # Example
zeta_4_num = sp.lambdify([cba], zeta_4_example, modules='numpy')

zeta_5_example = zeta_5_example.subs({MH:0.6/0.25, beta:np.pi/4}) # Example
zeta_5_num = sp.lambdify([cba], zeta_5_example, modules='numpy')

zeta_6_example = zeta_6_example.subs({MH:0.6/0.25, beta:np.pi/4}) # Example
zeta_6_num = sp.lambdify([cba], zeta_6_example, modules='numpy')

x_vals = np.linspace(-0.9999, 0.9999, 400)
y_vals_1 = zeta_1_num(x_vals)
y_vals_3 = zeta_3_num
y_vals_4 = zeta_4_num(x_vals)
y_vals_5 = zeta_5_num(x_vals)
y_vals_6 = zeta_6_num(x_vals)

plt.plot(x_vals,y_vals_1, label=r'$Z_1$')
plt.plot(x_vals,y_vals_3, label=r'$Z_3$')
plt.plot(x_vals,y_vals_4, label=r'$Z_4$')
plt.plot(x_vals,y_vals_5, label=r'$Z_5$')
plt.plot(x_vals,y_vals_6, label=r'$Z_6$')
plt.xlabel(r'cos($\beta-\alpha$)', fontsize=20)
plt.ylabel(r'$Z_i$', fontsize=20)
plt.xlim(-1, 1)
plt.grid()
plt.legend()
plt.show()
```

```python
lambda_1_example = lambda_1.subs({m:0, Y2:0, MHp:MH, MA:MH})  # Example
lambda_2_example = lambda_2.subs({m:0, Y2:0, MHp:MH, MA:MH})  # Example
lambda_3_example = lambda_3.subs({m:0, Y2:0, MHp:MH, MA:MH})  # Example
lambda_4_example = lambda_4.subs({m:0, Y2:0, MHp:MH, MA:MH})  # Example
lambda_5_example = lambda_5.subs({m:0, Y2:0, MHp:MH, MA:MH})  # Example

lambda_1_example = lambda_1_example.subs({MH:1, beta:np.pi/4}) # Example
lambda_1_num = sp.lambdify([cba], lambda_1_example, modules='numpy')

lambda_2_example = lambda_2_example.subs({MH:1, beta:np.pi/4}) # Example
lambda_2_num = sp.lambdify([cba], lambda_2_example, modules='numpy')

lambda_3_example = lambda_3_example.subs({MH:1, beta:np.pi/4}) # Example
lambda_3_num = sp.lambdify([cba], lambda_3_example, modules='numpy')

lambda_4_example = lambda_4_example.subs({MH:1, beta:np.pi/4}) # Example
lambda_4_num = sp.lambdify([cba], lambda_4_example, modules='numpy')

lambda_5_example = lambda_5_example.subs({MH:1, beta:np.pi/4}) # Example
lambda_5_num = sp.lambdify([cba], lambda_5_example, modules='numpy')

x_vals = np.linspace(-0.9999, 0.9999, 400)
y_vals_1 = lambda_1_num(x_vals)
y_vals_2 = lambda_2_num(x_vals)
y_vals_3 = lambda_3_num(x_vals)
y_vals_4 = lambda_4_num(x_vals)
y_vals_5 = lambda_5_num(x_vals)

plt.plot(x_vals,y_vals_1, label=r'$\lambda_1$')
plt.plot(x_vals,y_vals_2, label=r'$\lambda_2$')
plt.plot(x_vals,y_vals_3, label=r'$\lambda_3$')
plt.plot(x_vals,y_vals_4, label=r'$\lambda_4$')
plt.plot(x_vals,y_vals_5, label=r'$\lambda_5$')
plt.xlabel(r'cos($\beta-\alpha$)', fontsize=20)
plt.ylabel(r'$\lambda_i$', fontsize=20)
plt.xlim(-1, 1)
plt.grid()
plt.legend()
plt.show()
```

```python
Y2_value = 750 # GeV
vev = 246 # GeV # vev of the SM higgs boson
beta_value = np.pi/4

lambda_1 = lambda_1.subs({m:125/vev, Y2:Y2_value**2/vev**2, MHp:MH, MA:MH})
lambda_2 = lambda_2.subs({m:125/vev, Y2:Y2_value**2/vev**2, MHp:MH, MA:MH})
lambda_3 = lambda_3.subs({m:125/vev, Y2:Y2_value**2/vev**2, MHp:MH, MA:MH})
lambda_4 = lambda_4.subs({m:125/vev, Y2:Y2_value**2/vev**2, MHp:MH, MA:MH})
lambda_5 = lambda_5.subs({m:125/vev, Y2:Y2_value**2/vev**2, MHp:MH, MA:MH})
```

```python
X = np.linspace(-0.9999, 0.9999, 5000)
Y = np.linspace(0, 16, 5000)
x, y = np.meshgrid(X, Y)

lambda_1 = lambda_1.subs({beta:beta_value})
lambda_1_num = sp.lambdify([cba,MH], lambda_1, modules='numpy')
lambda_1 = lambda_1_num(x,y)

lambda_2 = lambda_2.subs({beta:beta_value})
lambda_2_num = sp.lambdify([cba,MH], lambda_2, modules='numpy')
lambda_2 = lambda_2_num(x,y)

lambda_3 = lambda_3.subs({beta:beta_value})
lambda_3_num = sp.lambdify([cba,MH], lambda_3, modules='numpy')
lambda_3 = lambda_3_num(x,y)

lambda_4 = lambda_4.subs({beta:beta_value})
lambda_4_num = sp.lambdify([cba,MH], lambda_4, modules='numpy')
lambda_4 = lambda_4_num(x,y)

lambda_5 = lambda_5.subs({beta:beta_value})
lambda_5_num = sp.lambdify([cba,MH], lambda_5, modules='numpy')
lambda_5 = lambda_5_num(x,y)

rho_3 = lambda_3 / np.sqrt(np.abs(lambda_1*lambda_2))
rho_4 = lambda_4 / np.sqrt(np.abs(lambda_1*lambda_2))
rho_5 = np.abs(lambda_5 / np.sqrt(np.abs(lambda_1*lambda_2)))

rho_3_eff = rho_3 + np.minimum(rho_4-abs(rho_5),0)



# Restrictions

restriction_1 = lambda_1 > 0
restriction_2 = lambda_2 > 0

restriction_3 = rho_3_eff + 1 > 0  # The condition obtanied in the Master's Thesis.
restriction_4 = lambda_1*lambda_2 - lambda_3**2 > 0  # Positive definite matrix condition.

# Additional restrictions in order to have a positive definite matrix. Due to the fact that these two restrictions might be too strong, we will not include them in the final plot.

# restriction_5 = lambda_5 > 0   
# restriction_6 = lambda_5**2 - lambda_4**2 > 0   

#########################################################################################################################################################################################

# Unitarity conditions


Ginzburg_Lambda_even_21p = 1/2 * (lambda_1 + lambda_2 + np.sqrt((lambda_1 - lambda_2)**2 + 4*np.abs(lambda_5)**2))
Ginzburg_Lambda_even_21m = 1/2 * (lambda_1 + lambda_2 - np.sqrt((lambda_1 - lambda_2)**2 + 4*np.abs(lambda_5)**2))
Ginzburg_Lambda_even_01p = 1/2 * (lambda_1 + lambda_2 + np.sqrt((lambda_1 - lambda_2)**2 + 4*lambda_4**2))
Ginzburg_Lambda_even_01m = 1/2 * (lambda_1 + lambda_2 - np.sqrt((lambda_1 - lambda_2)**2 + 4*lambda_4**2))
Ginzburg_Lambda_even_00p = 1/2 * (3*(lambda_1+lambda_2) + np.sqrt(9*(lambda_1-lambda_2)**2+4*(2*lambda_3+lambda_4)**2))
Ginzburg_Lambda_even_00m = 1/2 * (3*(lambda_1+lambda_2) - np.sqrt(9*(lambda_1-lambda_2)**2+4*(2*lambda_3+lambda_4)**2))
Ginzburg_Lambda_odd_21 = lambda_3 + lambda_4
Ginzburg_Lambda_odd_20 = lambda_3 - lambda_4
Ginzburg_Lambda_odd_01p = lambda_3 + np.abs(lambda_5)
Ginzburg_Lambda_odd_01m = lambda_3 - np.abs(lambda_5)
Ginzburg_Lambda_odd_00p = lambda_3 + 2*lambda_4 + 3*np.abs(lambda_5)
Ginzburg_Lambda_odd_00m = lambda_3 + 2*lambda_4 - 3*np.abs(lambda_5)

Ginzburg_restriction = (Ginzburg_Lambda_even_21p < 8*np.pi) & (Ginzburg_Lambda_even_21m < 8*np.pi) & (Ginzburg_Lambda_even_01p < 8*np.pi) & (Ginzburg_Lambda_even_01m < 8*np.pi) & (Ginzburg_Lambda_even_00p < 8*np.pi) & (Ginzburg_Lambda_even_00m < 8*np.pi) & (Ginzburg_Lambda_odd_21 < 8*np.pi) & (Ginzburg_Lambda_odd_20 < 8*np.pi) & (Ginzburg_Lambda_odd_01p < 8*np.pi) & (Ginzburg_Lambda_odd_01m < 8*np.pi) & (Ginzburg_Lambda_odd_00p < 8*np.pi) & (Ginzburg_Lambda_odd_00m < 8*np.pi)
```

```python
plt.figure(figsize=(25,10))

plt.subplot(1, 2, 1)

plt.contourf(y, x, restriction_1, colors=['lightgray','goldenrod'])
plt.contour(y, x, restriction_1, levels=[0.5], colors='black', linewidths=1)
plt.xlabel(r'$M_{H}/vev$', fontsize=20)
plt.ylabel(r'$\cos(\beta-\alpha)$', fontsize=20)
legend_patch = ppt.Patch(color='goldenrod', label='Valid region')
plt.legend(handles=[legend_patch], loc=2, fontsize=25)
plt.title(r'$\lambda_1 > 0$', fontsize=20)

plt.subplot(1, 2, 2)

plt.contourf(y, x, restriction_2, colors=['lightgray','goldenrod'])
plt.contour(y, x, restriction_2, levels=[0.5], colors='black', linewidths=1)
plt.xlabel(r'$M_{H}/vev$', fontsize=20)
plt.ylabel(r'$\cos(\beta-\alpha)$', fontsize=20)
legend_patch = ppt.Patch(color='goldenrod', label='Valid region')
plt.legend(handles=[legend_patch], loc=3, fontsize=25)
plt.title(r'$\lambda_2 > 0$', fontsize=20)

plt.show()
```
